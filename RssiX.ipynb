```python
def getXY(distancia, angulo):
    x = distancia*(math.cos(math.radians(angulo)))
    y = distancia*(math.sin(math.radians(angulo)))
    return x,y
```

```python
def Distancia_Euclidiana(xA,yA,xB,yB):
    distAB = math.sqrt(((xA-xB)**2) + ((yA-yB)**2))
    return distAB
```

```python
def Calcula_Distancia():
    VetorDistancia = []
    for i in range(len(y_pred)):
        pred = y_pred[i]
        #dado a categoria, busco a qual distancia e angulo ela pertence
        vetor = rssi[rssi.Label == pred].values[0:1,3:5] #pegando uma linha contendo distancia e angulo
        DistPredX = vetor[0,0] # distancia
        AnguloPred = vetor[0,1] # angulo
        PosPredX, PosPredY = getXY(DistPredX,AnguloPred)
        
        test = y_test[i]
        vetor2 = rssi[rssi.Label == test].values[0:1,3:5] #pegando uma linha contendo distancia e angulo
        DistTestX = vetor2[0,0] # distancia
        AnguloTest = vetor2[0,1] # angulo
        PosTestX, PosTestY = getXY(DistTestX,AnguloTest)
        
        VetorDistancia.append(Distancia_Euclidiana(PosPredX, PosPredY, PosTestX,PosTestY))
        print("Test - Distancia e Angulo\n", vetor2)
        print("Pred - Distancia e Angulo\n", vetor)
        print("Distancia em Metros entre eles: \n", VetorDistancia[i])
        print()
    #return VetorDistancia        
        
        
        
        
```
